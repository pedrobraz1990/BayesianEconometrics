# Numerical Methods for Bayesian Inference
Recall that once you have a Bayesian Statistical model, i.e., a prior and a likelihood the posterior is proportional to its multiplication

$$ p(\theta|x)=\frac{f(x|\theta)\pi(\theta)}{m(x)}\propto f(x|\theta)\pi(\theta)$$

As it is probably clear by know most of the Baysian Inference comes from solving some form of problem involving an integral as

$$I(g) = \int g(\theta)p(\theta|x)\mathrm{d}\theta $$

For instance:
1. If we take $g(\theta)=1$ we have that $I(g)=k m(x)$, which is the marginal distribution of the data that can be used to normalize the posterior (integrate to unity);
2. If we take $g(\theta)=\theta$ we have the posterior mean. In fact if take $g(\theta)$ itself we have $I(g)=\mathbb{E}(g(\Theta)|X=x)$
3. If we take $g(\theta)=\theta\theta'$ and $\Theta$ is zero mean, $I(g)$ becomes the covariance matrix of $\Theta|X=x$
3. If we take $g(\theta)=1\{\theta\leq k\}$ and $I=1/2$ and solve for $k$
we obtain the median of the posterior;
4. If we take $g(\theta)=1\{\theta\in C\}$ and $I=1-\alpha$ and solve for $C$ we can obtain a $1-\alpha$ credible set;

The marginalization with regards to a subset of $\theta$ can also be accommodated in this framework since the marginal of $\theta_1$ is given by:

$$p(\theta_1|x) = \int p(\theta|x)\mathrm{d}\theta_2 = \int p(\theta_1|\theta_2,x)p(\theta_2|x)\mathrm{d}\theta_2 $$
where $g(\theta_2,x)=p(\theta_1|\theta_2,x)$

For the model selection framework if we set 

$$I_j = \int g(\theta_j)p(\theta|x,\mathcal{M}_j)\mathrm{d}\theta_j $$

and choose $g(\theta_j) =1$, $I_j=m(x|\mathcal{M}_j)$ becomes the marginal distribution of the data given the model $\mathcal{M}_j$. Which can be used to compare the $J$ available models by the posterior odds for instance.

Even the forecasting procedure can be casted in this framework by considering a slight change. Recall that that given a predictive model $h(y|\theta,x)$ the predictive density is given by

$$p(y|x)=\int h(y|\theta,x)p(\theta|x)\mathrm{d}\theta, $$
where, once again, $g(\theta,x) = h(y|\theta,x)$.

Aside from conjugate priors, most of those integrals can not be solved analytically either because there is no closed form or because the algebra would be prohibited. There three basically three approaches when it comes to solve a integral numerically

1. Deterministic Methods
2. Non-iterative Simulation Methods
3. Iterative Simulation Methods

## Deterministic Methods

Those methods were not originally motivated by Statistics. They mimic the Riemann integral by constructing a grid in the sample space of the posterior. You end up suffering from what is called the curse of dimensionality as it becomes increasing difficult to get a good approximation of the integral as the dimension of the posterior increases.  Perform poorly even for reasonably small models (up to 5 parameters). 


## Monte Carlo Integration

To take advantage of the fact that our integrand is not a ordinary function but a density, an integration through a stochastic simulation was proposed known as the **Monte Carlo (MC) Integral**. The idea comes from the Law of Large Number since we know that

$$\frac{1}{n}\sum_{i=1}^n X_i\approx\mathbb{E}(X)=\int x f(x)\mathrm{d}x $$

Thus given a sample $\theta_1,\dots\theta_n$ from $p(\theta|x)$ we expect that 

$$\widehat{I} = \frac{1}{n}\sum_{i=1}^n g(\theta_i)\approx\mathbb{E}(g(\Theta|X=x))=\int g(\theta) p(\theta|x)\mathrm{d}\theta $$

$\widehat{I}(x)$ is sometimes called the MC estimator and enjoying several "frequentist"properties under mild conditions:
* Unbiasedness: $\mathbb{E} \widehat{I}(x) = I(x)$;
* Variance: $\Omega\equiv\mathbb{V}\widehat{I}(x)= (\mathbb{E}g(\Theta)^2 - I^2)/n$;
* Strong Consistency: $\widehat{I}(x) \overset{a.s.}{\rightarrow}  I(x)$;
* Asymptotically Normality: $\sqrt{n}(\widehat{I}(x) - I)\overset{d}{\rightarrow} N(0,\Omega)$

Hence the MC error in of order $O(n^{-1/2})$ and, unlike the frequentist world, $n$ is controlled by the experimenter so you can increase $n$ arbitrary to have a virtually a error-free estimation.

```r
# Monte Carlo integration Example
# Code to numeric evalute the Bayesian Binomial-Beta Conjugate
library(ggplot2)
library(repr)

s =100000
n = 10
x = 8
a = x+1/2
b = n-x+1/2
sampled = rbeta(s,a,b)

# Ploting
options(repr.plot.width=4, repr.plot.height=4)
true = dbeta(sampled,a,b)
data = data.frame(sampled,true)
ggplot(data, aes(x = sampled)) + 
  geom_histogram(aes(y=..density..),binwidth = 0.02)+
  geom_line(aes(y = true, colour = "Posterior")) +
  theme(legend.position = c(0.8, 0.8)) +
  theme(legend.title=element_blank())

Numeric_MC = c(mean(sampled),median(sampled),sampled[which(true==max(true))],var(sampled))
Real = c(a/(a+b), (a-1/3)/(a+b-2/3), (a-1)/(a+b-2), a*b/((a+b)^2*(a+b+1)))
tab = data.frame(Numeric_MC,Real)
row.names(tab)= c("Mean","Median","Mode","Variance")
tab
```

The Monte Carlo Integration is completely dependent on a sample drawn from the posterior, which can be computationally inefficient and/or costly specially complex models. We now present two alternative to deal with this problem:

### Rejection Sampling

The exist a fundamental simple idea behind the accept-reject methodology. For a given density $p(x)$ we can trivially write

$$p(x)=\int_0^{p(x)}\mathrm{d}u $$

thus, $p(x)$ appears as the marginal density in $X$ of the joint distribubution of

$$(X,U)\sim \text{Uniform}\{(x,u):0\leq u\leq p(x)\} $$

Unfortunately is not often straight forward to simulate the pair $(X,U)$ without being able to draw from $p(x)$ itself but you could simulate the pair $(X,U)$ on a bigger set, where simulation is easier and take take the pair if the constrain is satisfied.

For instance in the one-dimensional case suppose that $X\sim p(x)$ has support $[a,b]$ and p(x) is bounded by $m$, say. We can then simulate the pair $(Y,U)\sim \text{Uniform}(0\leq u\leq m)$ by simulating $Y\sim U(a,b)$ and $U\sim U(0,m)$ and take only the pairs $(y,u)$ such that the constrain $0\leq u\leq p(y)$ is satisfied. This way the distribution of the accepted $Y$, call it $X$ is $p$ since:

$$P(X\leq x) = P(Y\leq x|U\leq p(Y)) = \frac{\int_a^x\int_0^{p(y)}\mathrm{d}u\mathrm{d}y}{\int_a^b\int_0^{p(y)}\mathrm{d}u\mathrm{d}y} = \int_a^{x}p(y)\mathrm{d}y$$

Moreover, it is easy to see that the probability of acceptance of a given simulation of the box $[a,b]\times[0,m]$ is given by

$$P(\text{Accept})=P(U<p(Y))=\mathbb{E}\left[P(U<p(Y)|Y\right] = \frac{1}{m}\mathbb{E}\left[P(Y)\right] = \frac{1}{m}\int_a^b p(y)\mathrm{d}y=\frac{1}{m} $$

The argument can be further generalize to when the larger set is no longer a box, as long as simulating uniformly over this larger set is feasible. This generalization is important to accommodate distributions with unbounded support and/or unbounded density. The larger set is then of the form

$$A = \{(y,u):0\leq u \leq m(y)\} $$

with the constrain that $p(x)\leq m(x)$ (the set is actually larger than the area under the graph of $p(x)$. 

Notice that because of this constrain $m(y)$ cannot be a density (does not integrate to unity). In practice, we choose an easy-to-sample proposal distribution $q(x)$ that satisfies the "envelope condition"

$$p(x) â‰¤ Mq(x),\quad M <\infty $$

Hence we can simulate $Y\sim q$ and then $U|Y=y\sim U(0,Mq(y)$ to create the pairs $(y,u)$ and only accept those tha satisfies $u<p(y)$. Recall that the event of acceptance

$$\{U<p(Y)\}=\left\{\frac{U}{Mq(Y)}<\frac{p(Y)}{Mq(Y)}\right\}\equiv\left\{U^*<\frac{p(Y)}{Mq(Y)}\right\}$$

We can sample from a distribution $p(x)$, which is known up to a proportionality constant, by sampling from another easy-to-sample proposal distribution $q(x)$ that satisfies the "envelope condition"



using the accept/reject procedure describe below 

---
**Rejection Sampling Algorithm**

Set $i=1$ and while $i<n$
1. Draw $x^{(i)}\sim q(x)$ and $u\sim U(0,1)$
2. If $u<\frac{p(x^{(i)})}{M q(x^{(i)})}$, accept $x^{(i)}$ and $i=i+1$, otherwise go back to 1.

---

This simple method suffers from severe limitations. It is not always possible to bound $p(x)/q(x)$ with a reasonable constant $M$ over the whole support of $x$. If M is too large, the acceptance probability

  $$P(x \text{ accepted}) = P\left(u<\frac{p(x)}{M q(x)}\right) = \frac{1}{M}$$
  
will be too small. This makes the method impractical in high-dimensional scenarios.

Ideally we want to choose $M$ as close to one as possible since we have a geometric waiting time hence the expected number of trials before getting one accepted sample is $M$. To keep $p/q$ uniformly bounded we have $M\geq 1$ which equality only if $p=q$ and have $q$ with thicker tails than $p$ hence there is no way, for instance, to use a Normal to simulate Cauchy; however the reverse works quite well.

Assume that the posterior density $p(x)$ is only known up to a proportionality constant (which is the case in most situations). Therefore, $p^*(x) = f(x)\pi(x)$ is available where $f$ is the likelihood and $\pi$ is the prior for $x$ but $p(x) = p^*(x)/k$ is not because the value of $k>0$ is unknown. The rejection sampling method work just fine since we can always absorb $k$ into $M$ a define $M^*=kM$.

```r
# Rejection Sampling Example
# Code to numeric evalute the Bayesian Binomial-Beta Conjugate

# INPUTS
n = 100000
x = runif(n)
q = dunif(x,0,1)
p = dbeta(x, 3,6)
M = 1

# OUTPUT
c = p/q/M
sampled = x[runif(n) < c]
ratio = length(sampled)/n

#Plotting
data = data.frame(x,p,q)
ggplot(data.frame(sampled), aes(x = sampled)) + 
  geom_histogram(aes(y=..density..),binwidth = 0.02)+
  geom_line(data=data, aes(x = x, y = p, colour = "Posterior (p)")) +
  geom_line(data=data, aes(x = x, y = M*q, colour = "Proposal (M*q)")) +
  ggtitle(paste("Rejection Sampling - Acceptance Ratio", ratio)) +
  xlab("Sampled") +
  theme(legend.position = c(0.85, 0.95),
        legend.title=element_blank(),
        axis.title.y=element_blank(),
        plot.title = element_text(hjust = 0.5),
        aspect.ratio=1
       ) 
```

### Importance Sampling

If we are mostly interested in the value of the functional $I(g)$ instead of the posterior itself we can calculate an estimator of $I(g)$ without actually sampling from $p(x)$. Even better, most of the times there actually a more efficient (smaller variance) way to estimate $I(g)$. We present the general idea through an example

***EXAMPLE:*** Supose you are interested in the probability of a Cauchy(0,1) random variable le greater than 2. Call this probability $p$ and we know it is $0.15$.

$$p=\int_2^{\infty}\frac{1}{\pi(1+x^2)}\mathrm{d}x $$

For a fix number of simulations for the Cauchy(0,1) $X_1,\dots, X_n$ let's consider the following estimators:

* Plain MC: $\widehat{p}_1=\frac{1}{n}\sum_{i=1}^n 1\{X_i>2\}$
* MC + Symmetric: $\widehat{p}_2=\frac{1}{2n}\sum_{i=1}^n 1\{|X_i|>2\}$
 
If we consider $h(X)\equiv \frac{2}{\pi(1+ X^2)}$, where $X\sim U(0,2)$, we can rewrite $p$ as

$$p = \frac{1}{2}-\int_0^2\frac{1}{\pi(1+x^2)}\mathrm{d}x = \frac{1}{2}-\int_0^2h(X)\frac{1}{2}\mathrm{d}x = \frac{1}{2}- \mathbb{E}_X(h(X))$$, which suggests the following estimator 

* $\widehat{p}_3=\frac{1}{2} -\frac{1}{n}\sum_{i=1}^n h(U_i)$






```r
# Comparing Cauchy Estimators

n = 10
S = 1000
P = matrix(0,S,3)
for (i in 1:S) {
x = rcauchy(n)
P[i,1] = mean(x>2)
P[i,2] = sum(abs(x)>2)/(2*n)
u = runif(n,0,2)
h = 2/(pi*(1+u^2))
P[i,3] = 0.5-mean(h)    
    }
Mean = colMeans(P)
Variance = apply(P,2,var)

tab = rbind(Mean,Variance)
tab
```

From the example notice that you could potentially convert the expectation (taken initially with respect to the Cauchy distribution) to any other distribution (the uniform as the case for the third estimator). The question that naturally arises is that what are the constrains? Is there a lower bound for the variance of the estimators we can obtain in practice?

Let us introduce, again, an arbitrary importance proposal distribution $q(x)$ such that its support includes the support of $p(x)$. Then we can rewrite $I ( g )$ as follows

$$I(g) = \int g(x)p(x)\mathrm{d}x  = \int g(x)\frac{p(x)}{q(x)}q(x)\mathrm{d}x \equiv \int g(x)w(x)q(x)\mathrm{d}x =\mathbb{E}_q(g(X)w(X)) $$

where $w(x)$ is known as the **importance weight**.

Even though we asked for the support of $q(x)$ to be included in the support of $p(x)$, it is enough to have for a given $g(x)$

$$q(x)>0 \text{ whenever } g(x)p(x)\neq 0 $$

However if you want to use a single $q(x)$ for several choices of $g(x)$  to have the support of $q(x)$ to be included in the support of $p(x)$ is sufficient.

It is natural to wounder what happen when $q(x)=0$ in the denominator. In practice it will never happen since we are sampling for $q(x)$ and if you have a sample $x^{(i)}$ it must be the case that $q(x^{(i)})>0$. From the theoretical point of view, $q(x)=0$ is it also not a problem as long $g(x)p(x)=0$ as well.

Consequently, if one can sample $n$ iid observations $\{X_i\}_{i=1}^n$ from $q(x)$ and evaluate $w(X_i)$, a possible Monte Carlo
estimate of $I(g)$ is given by

$$\widehat{I}_n(g)=\frac{1}{n}\sum_{i=1}^n g(X_i)w(X_i),\quad X_i\sim q $$

In practice we are making a multiplicative adjustment factor $w(x)$ to $g(x)$ to compensate for sampling from $q(x)$ instead of $p(x)$. Of course it only makes sense if it is easier to sample from $q$ then $p$. Even though there is an argument towards importance sampling even when sampling from $p(x)$ is feasible. 

The variance of our importance sampling estimator is given by

$$\mathbb{V}_q \widehat{I}_n(g) = \frac{1}{n}\mathbb{V}_q \left(g(X_i)w(X_i)\right) = \frac{1}{n}\left(\mathbb{E}_qg^2(X_i)w^2(X_i)-I^2(g)\right)$$

The second term does not depend upon $q(x)$ so we concentrate on the first one and bound it by Jensen Inequality as

$$\mathbb{E}_qg^2(X_i)w^2(X_i)\geq \left(\mathbb{E}_q |g(X_i)|w(X_i)\right)^2 =\left(\int |g(x)|p(x)\mathrm{d}x\right)^2,$$

which is a lower bound that is independent of the choice of $q(x)$. It is straight forward to see that the equality will hold with the  **optimal importance distribution** given by

$$q^*(x)=\frac{|g(x)|p(x)}{\mathbb{E}_p |g(X)|}\propto |g(x)|p(x).$$

The optimal proposal is not very useful in the sense that it is not easy to sample from $| g (x )| p(x )$. However, it tells us that high sampling efficiency is achieved when we focus on sampling from $p(x)$ in the important regions where $| g (x)|p(x)$ is relatively large; hence the name importance sampling.

This result implies that importance sampling estimates can be **super-efficient**. That is, for a a given function $g(x)$, it is possible to find a distribution $q(x)$ that yields an estimate with a lower variance than when using a perfect Monte Carlo method, i.e. with $q(x) = p(x)$.

When the normalizer of the posterior is known, such that we only have $p^*(x)$, the importance sampling method can still be applied with a slightly  modification to the estimator called the **self-normalized importance sampling estimator**

$$\widehat{I}_n(g)=\frac{\sum_{i=1}^n g(X_i)w(X_i)}{\sum_{i=1}^n w(X_i)},\quad X_i\sim q $$
